## Part 3: Load Data Using Polars

```python
# Rename 'Date_' to 'Date'
data.rename(columns={'Date_': 'Date'}, inplace=True)

# Now proceed with the melt operation
df_long = data.melt(id_vars=["Date"], 
                    value_vars=[col for col in data.columns if col != "Date"],
                    var_name="company_variable", 
                    value_name="value")

# Split 'company_variable' into 'company_name' and 'metric' (e.g., Open, Close, High, Low, Volume)
df_long[['metric', 'company_name']] = df_long['company_variable'].str.split('_', expand=True)

# Filter for the relevant metrics (Open, High, Low, Close, Volume)
df_long = df_long[df_long['metric'].isin(['Open', 'High', 'Low', 'Close', 'Volume'])]

# Drop the 'company_variable' column
df_long = df_long.drop(columns='company_variable')

# Clean the company_name column in Pandas before converting to Polars
df_long['company_name'] = df_long['company_name'].str.replace('"', '', regex=False)

# Pivot the data to have one row per Date and company_name
df_pivoted = df_long.pivot_table(index=['Date', 'company_name'], columns='metric', values='value', aggfunc='first')

# Reset index and reorder the columns
df_pivoted = df_pivoted.reset_index()
df_pivoted = df_pivoted[['Date', 'company_name', 'Open', 'High', 'Low', 'Close', 'Volume']]

# Clean company_name column again after pivoting
df_pivoted['company_name'] = df_pivoted['company_name'].str.replace('"', '', regex=False)

# Convert to Polars DataFrame
df_polars = pl.from_pandas(df_pivoted)

# Save as Parquet for efficiency
df_polars.write_parquet("tech_stocks_data.parquet", compression="snappy")
```

## Part 4: Feature Engineering
We will compute technical indicators (SMA, RSI, MACD, Bollinger Bands, and rolling volatility):

```python
# Convert to pandas for calculations
df_pd = df_polars.to_pandas()
df_pd.reset_index(inplace=True)

# Function to calculate RSI
def compute_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# Initialize a list to store each company's DataFrame
group_list = []

# Loop through each company group
for company, group in df_pd.groupby("company_name"):    
    group = group.copy()  # Prevent SettingWithCopyWarning 
    group["SMA_10"] = group["Close"].rolling(window=10).mean()   
    group["SMA_30"] = group["Close"].rolling(window=30).mean()    
    group["RSI_14"] = compute_rsi(group["Close"])   
    group["MACD"] = group["Close"].ewm(span=12, adjust=False).mean() - group["Close"].ewm(span=26, adjust=False).mean()    
    group["Bollinger_Upper"] = group["Close"].rolling(window=20).mean() + (group["Close"].rolling(window=20).std() * 2)    
    group["Bollinger_Lower"] = group["Close"].rolling(window=20).mean() - (group["Close"].rolling(window=20).std() * 2)    
    group["Rolling_Volatility"] = group["Close"].pct_change().rolling(window=20).std()    
    group_list.append(group)

# Concatenate all groups back together
df_pd = pd.concat(group_list, axis=0, ignore_index=True)

# Reorder/Select the desired columns
df_pd = df_pd[["Date", "company_name", "Open", "High", "Low", "Close", "Volume","SMA_10", "SMA_30", "RSI_14", "MACD", "Bollinger_Upper",
                "Bollinger_Lower", "Rolling_Volatility"]]

# Convert back to Polars
df = pl.from_pandas(df_pd)

# Save enriched data
df.write_parquet("tech_stocks_enriched.parquet", compression="snappy")

# Verify results
df_pd.head()
```

### 6.1 Long Short-Term Memory Model (LSTM)
- Non Linear Model

```python
from tensorflow.keras.layers import Input

# Ensure Date is in datetime format
df_pd["Date"] = pd.to_datetime(df_pd["Date"])

# Pivot data for LSTM: each company's Close becomes a column, Date as index
df_lstm = df_pd.pivot(index='Date', columns='company_name', values='Close')
df_lstm.dropna(inplace=True)

# Scale data
scaler_lstm = MinMaxScaler(feature_range=(0, 1))
df_scaled_lstm = scaler_lstm.fit_transform(df_lstm)

# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 60
X_lstm, y_lstm = create_sequences(df_scaled_lstm, seq_length)

# Train-test split
train_size_lstm = int(len(X_lstm) * 0.8)
X_train_lstm, X_test_lstm = X_lstm[:train_size_lstm], X_lstm[train_size_lstm:]
y_train_lstm, y_test_lstm = y_lstm[:train_size_lstm], y_lstm[train_size_lstm:]

# Build LSTM model
model_lstm = Sequential()
model_lstm.add(Input(shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model_lstm.add(LSTM(100, return_sequences=True))
model_lstm.add(Dropout(0.3))
model_lstm.add(LSTM(100, return_sequences=False))
model_lstm.add(Dropout(0.3))
model_lstm.add(Dense(y_train_lstm.shape[1]))

model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model_lstm.fit(X_train_lstm, y_train_lstm, epochs=100, batch_size=64, validation_data=(X_test_lstm, y_test_lstm), verbose=1)

# Predict
predictions_lstm = model_lstm.predict(X_test_lstm)
predictions_lstm = scaler_lstm.inverse_transform(predictions_lstm)

# Evaluate
mae_lstm = np.mean(np.abs(predictions_lstm - scaler_lstm.inverse_transform(y_test_lstm)))
rmse_lstm = np.sqrt(np.mean((predictions_lstm - scaler_lstm.inverse_transform(y_test_lstm))**2))
print(f"LSTM MAE: {mae_lstm}, RMSE: {rmse_lstm}")
```

```python
import matplotlib.dates as mdates

# Plot predictions vs actual for each company
for i, company in enumerate(df_lstm.columns):
    plt.figure(figsize=(10, 4))
    plt.plot(df_lstm.index[train_size_lstm + seq_length:], scaler_lstm.inverse_transform(y_test_lstm)[:, i], label=f'Actual {company}')
    plt.plot(df_lstm.index[train_size_lstm + seq_length:], predictions_lstm[:, i], label=f'Predicted {company}', color='red')
    plt.title(f"LSTM Forecast for {company}")
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    
    # Format x-axis to show months and years
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### 6.2 Light Gradient Boosting Machine Model (LightGBM)
This approach allows you to incorporate explanatory variables into cross section analysis effectively.
- Features: Use technical indicators and other relevant variables as features.
- Evaluation: Provides MAE and RMSE for model performance.

```python
#Copy dataframe

selected_columns = [
    "Date", "company_name", "Open", "High", "Low", "Close", "Volume",
    "SMA_10", "SMA_30", "RSI_14", "MACD", "Bollinger_Upper", "Bollinger_Lower",
    "Rolling_Volatility", "USD_to_ALL", "Dow_Jones_Index", "SP500_Index"
]

df_pd_safe = df_pd[selected_columns].copy()
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor

# Step 1: Create target column (next day's Close)
df_pd_safe["close_next"] = df_pd_safe.groupby("company_name")["Close"].shift(-1)

# Step 2: Drop rows with missing target
df_model_safe = df_pd_safe.dropna(subset=["close_next"]).copy()

# Step 3: Define feature columns (excluding identifiers and target)
feature_cols_safe = [
    col for col in df_model_safe.columns
    if col not in ["Date", "company_name", "Close", "close_next"]
]

# Step 4: Initialize empty lists to store train/test splits for each company
X_train_list = []
X_test_list = []
y_train_list = []
y_test_list = []

# Step 5: Split data by company (with random shuffle)
for company in df_model_safe["company_name"].unique():
    # Get data for the current company
    company_data = df_model_safe[df_model_safe["company_name"] == company]
    
    # Shuffle the data randomly
    company_data_shuffled = company_data.sample(frac=1, random_state=42)  # frac=1 means shuffle all rows
    
    # Split the data: 80% for training, 20% for testing
    X_train_company, X_test_company, y_train_company, y_test_company = train_test_split(
        company_data_shuffled[feature_cols_safe],
        company_data_shuffled["close_next"],
        test_size=0.2,
        random_state=42
    )
    
    # Append the results to the lists
    X_train_list.append(X_train_company)
    X_test_list.append(X_test_company)
    y_train_list.append(y_train_company)
    y_test_list.append(y_test_company)

# Step 6: Concatenate all the splits
X_train_safe = pd.concat(X_train_list)
X_test_safe = pd.concat(X_test_list)
y_train_safe = pd.concat(y_train_list)
y_test_safe = pd.concat(y_test_list)

# Step 7: Train LightGBM model
model_safe = LGBMRegressor()
model_safe.fit(X_train_safe, y_train_safe)

# Step 8: Evaluate model
y_pred_safe = model_safe.predict(X_test_safe)

# Step 9: Predict for all valid rows in df_pd_safe
df_pd_safe["close_next_pred"] = None
valid_rows_safe = df_pd_safe[feature_cols_safe].notnull().all(axis=1)
df_pd_safe.loc[valid_rows_safe, "close_next_pred"] = model_safe.predict(df_pd_safe.loc[valid_rows_safe, feature_cols_safe])
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Get unique company names
company_names = df_pd_safe['company_name'].unique()

# Loop through each company and plot its data
for company in company_names:
    # Filter data for the current company
    company_data = df_pd_safe[df_pd_safe['company_name'] == company]

    # Plot actual vs predicted close_next vs close_next_pred for the current company
    plt.figure(figsize=(10, 4))
    plt.plot(company_data['Date'], company_data['close_next'], label='Actual Close Next', color='blue')
    plt.plot(company_data['Date'], company_data['close_next_pred'], label='Predicted Close Next', color='red')
    plt.title(f"Close Next vs Predicted Close Next for {company}")
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()

    # Format x-axis to show only the start of each year
    plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Show ticks for each year
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # Show year format

    # Rotate x-ticks for better readability
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Show plot for the current company
    plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Step 1: Evaluate the refined model (MAE, RMSE)
mae_top_safe = mean_absolute_error(y_test_safe, y_pred_safe)
rmse_top_safe = np.sqrt(mean_squared_error(y_test_safe, y_pred_safe))

# Step 2: Print the evaluation metrics
print(f"Refined LightGBM MAE: {mae_top_safe:.4f}")
print(f"Refined LightGBM RMSE: {rmse_top_safe:.4f}")
```
